```python
import pandas as pd
```

```python
data = pd.read_excel("ENVISON_DATA_PS3.xlsx")
data
```

```python
# Map to actual column names and keep only specified features
columns_to_keep = [
    'DisNo.',  # event_id
    'Disaster Type',  # disaster_type
    'Disaster Subtype',  # disaster_subtype
    'Location',  # location
    'Latitude',  # latitude
    'Longitude',  # longitude
]

# Add date-related columns (we'll combine Start Year, Month, Day)
columns_to_keep.extend(['Start Year', 'Start Month', 'Start Day'])

# Add remaining columns that exist
additional_columns = [
    'Total Affected',  # affected_population
    "Total Damage ('000 US$)",  # estimated_economic_loss_usd
]

columns_to_keep.extend(additional_columns)

# Filter the dataset
data = data[columns_to_keep]

# Rename columns to match requested names
data = data.rename(columns={
    'DisNo.': 'event_id',
    'Disaster Type': 'disaster_type',
    'Disaster Subtype': 'disaster_subtype',
    'Location': 'location',
    'Latitude': 'latitude',
    'Longitude': 'longitude',
    'Total Affected': 'affected_population',
    "Total Damage ('000 US$)": 'estimated_economic_loss_usd'
})

data
```

```python
# Check for missing latitude/longitude values
print("Missing latitude values:", data['latitude'].isnull().sum())
print("Missing longitude values:", data['longitude'].isnull().sum())
print("\nRows with missing coordinates:")
missing_coords = data[data['latitude'].isnull() | data['longitude'].isnull()]
print(f"Total rows with missing coordinates: {len(missing_coords)}")
```

```python
# Install geopy for geocoding
# !pip install geopy

from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut, GeocoderServiceError
import time

# Initialize geocoder
geolocator = Nominatim(user_agent="disaster_data_geocoder")

def get_coordinates(location):
    """Get latitude and longitude for a location"""
    if pd.isna(location) or location == '':
        return None, None
    
    try:
        time.sleep(1)  # Respect rate limits
        loc = geolocator.geocode(location, timeout=10)
        if loc:
            return loc.latitude, loc.longitude
        return None, None
    except (GeocoderTimedOut, GeocoderServiceError):
        return None, None

# Fill missing coordinates
print("Filling missing coordinates...")
for idx, row in data[data['latitude'].isnull()].iterrows():
    if pd.notna(row['location']):
        lat, lon = get_coordinates(row['location'])
        if lat and lon:
            data.at[idx, 'latitude'] = lat
            data.at[idx, 'longitude'] = lon
            print(f"Updated {row['location']}: ({lat}, {lon})")

print("\nAfter filling:")
print("Missing latitude values:", data['latitude'].isnull().sum())
print("Missing longitude values:", data['longitude'].isnull().sum())
```
